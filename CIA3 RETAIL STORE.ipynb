```python
# Visualization done using the bar graph
import matplotlib.pyplot as plt
import seaborn as sns
features=['Gender','Branch','Customer type','Product line','Payment']
fig=plt.subplots(figsize=(15,25))
for i, j in enumerate(features):
    plt.subplot(4, 2, i+1)
    plt.subplots_adjust(hspace = 0.5)
    sns.countplot(x=j,data = data1)
    plt.xticks(rotation=90,fontsize=14)
    plt.title("")
```

```python
plt.figure(figsize=(14,6))
ax = sns.countplot(x = "Customer type", hue = "Branch", data = data1, palette= "rocket_r")
ax.set_title(label = "Customer type in different branch", fontsize = 25)
ax.set_xlabel(xlabel = "Branches", fontsize = 16)
ax.set_ylabel(ylabel = "Customer Count", fontsize = 16)
```

```python
plt.figure(figsize = (14,6))
ax = sns.countplot(x="Branch", hue = "Payment", data = data1, palette= "tab20")
ax.set_title(label = "Payment distribution in all branches", fontsize= 25)
ax.set_xlabel(xlabel = "People Count", fontsize = 16)
ax.set_ylabel(ylabel = "Payment method", fontsize = 16)
```

```python
plt.figure(figsize=(14,6)) 
ax = sns.boxplot(x="Branch", y = "Rating" ,data =data1, palette= "RdYlBu")
ax.set_title("Rating distribution between branches", fontsize = 25)
ax.set_xlabel(xlabel = "Branches", fontsize = 16)
ax.set_ylabel(ylabel = "Rating distribution", fontsize = 16)
```

```python
# Visualization done using the line graph
plt.figure(figsize=(14,6)) 
rating_vs_sales = sns.lineplot(x="Quantity", y= "Total", data=data1)
```

```python
# visualization done using the scatterplot.
plt.scatter(data1['Quantity'],data1['cogs'])
plt.xlabel("Quantity",fontsize=16)
plt.ylabel("COGS",fontsize=16)
plt.title("Quantity Vs COGS",fontsize=16,color="red")
plt.show()
```
